```python
import pandas as pd                 # pandas is a dataframe library
import matplotlib.pyplot as plt      # matplotlib.pyplot plots data
```

```python
df = pd.read_csv("./parkinsons.data")
df.head(5) 
# print(df.isnull().values.any()) # checking for null values
del df['name']  # Deleting the column name
del df['MDVP:Jitter(%)']  # Deleting the column because of corr
del df['MDVP:Shimmer(dB)']  # Deleting the column because of corr 
del df['MDVP:RAP']  # Deleting the column because of corr 
del df['Shimmer:APQ5']  # Deleting the column because of corr 
del df['Shimmer:APQ3']  # Deleting the column because of corr 
del df['Shimmer:DDA']  # Deleting the column because of corr 
del df['PPE']  # Deleting the column because of corr 
```

```python
def plot_corr(df, size=20):
    corr = df.corr()  # data frame correlation function
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)  # color code the rectangles by correlation value
    print(plt.xticks(range(len(corr.columns)), corr.columns))  # draw x tick marks
    print(plt.yticks(range(len(corr.columns)), corr.columns))  # draw y tick marks
    
```

```python
print(df.head(5))
plot_corr(df)
```
